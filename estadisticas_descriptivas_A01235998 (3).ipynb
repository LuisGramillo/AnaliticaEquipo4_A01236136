### **5. Basándose en la media, mediana y desviación estándar de cada variable cuantitativa, indica que conclusiones puedes entregar de los datos, escribe esto dentro del mismo notebook**

**Variable 'Pregnancies'**

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.displot(diabetes['Pregnancies'], bins=10)
plt.show()
print("Media:",diabetes['Pregnancies'].mean())
print("Mediana:",diabetes['Pregnancies'].median())
print("Desviación estándar:",diabetes['Pregnancies'].std())
```

*Esto significa que el promedio de embarazos en las personas encuestadas es de 3 a 4, y que este numero varia 3.4 con respecto a la media*
